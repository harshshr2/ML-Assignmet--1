```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

data_train=pd.read_csv('/data/training/car_train.csv')

X_train = data_train.iloc[:,:-1] 
Y_train = data_train.iloc[:,-1]


reg = LinearRegression()   # creating object of linear regression

reg = reg.fit(X_train,Y_train)

data_test=pd.read_csv('/data/test/car_test.csv')

Y_test = data_test["speed"].values.reshape(-1, 1)
Y_pred = reg.predict(Y_test)

rmse = np.sqrt(mean_squared_error(Y_test, Y_pred))

Y_pred_DF =pd.DataFrame(Y_pred)
data_fin = pd.concat([data_test.iloc[:,:-1], Y_pred_DF],axis=1)
data_fin.columns = ['id', 'dist']

print("RMSE =", rmse)
data_fin.to_csv("/code/car_prediction.csv")
```

```python
Y_pred = reg.predict(X)
plt.scatter(X,Y, color='#1835d2')
plt.plot(X,Y_pred, color = 'red')
```

```python
pd.DataFrame(Y_pred).reset_index()
```

```python
# Model Evaluation
from sklearn.metrics import r2_score

rmse = np.sqrt(mean_squared_error(Y, Y_pred))
r2 = r2_score(Y, Y_pred)   # built-in function score() indicates R-squared value

print("RMSE =", rmse)
print("R2 score =", r2)
```
